```python
cmp['Request_closing_time'] = cmp['Closed Date'] - cmp['Created Date']
cmp
```

```python
cmp['Request_closing_time_secs'] = cmp['Request_closing_time'].apply(lambda x:x.seconds)
cmp['Request_closing_time_secs'] = cmp['Request_closing_time_secs']
cmp = cmp[cmp['Request_closing_time_secs'].notnull()]
cmp.info()
```

```python
Comp_action_time = cmp[['Complaint Type','Request_closing_time_secs']]
Comp_action_time2 = Comp_action_time.groupby('Complaint Type')['Request_closing_time_secs'].mean().to_frame()
Comp_action_time2 = Comp_action_time2.sort_values('Request_closing_time_secs')
Comp_action_time2['Complaint Type'] = Comp_action_time2.index
txt = {'weight':'bold'}
plt.figure(figsize=(16,8))
sns.barplot(x = 'Complaint Type',y = 'Request_closing_time_secs',data=Comp_action_time2)
plt.title("Complaint type with avg resolve time",fontdict=txt)
plt.xlabel("Complaint Type",fontdict=txt,labelpad=30)
plt.ylabel("Avg request resolve time",fontdict=txt,labelpad=30)
plt.xticks(rotation=45)
plt.show
```

Conclusion 4: Animal in a park and posting advertisement complaints are responded faster
